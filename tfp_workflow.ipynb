```python
import tensorflow as tf
import tensorflow_probability as tfp
from matplotlib import pyplot as plt
tfd = tfp.distributions
import numpy as np
import pandas as pd
```

```python
normal = tfd.Normal(loc=0.0, scale=1.0)
bernoulli = tfd.Bernoulli(probs=0.7)
batched_bernoulli = tfd.Bernoulli(probs=[0.7,0.2,0.5])
exponential = tfd.Exponential(rate=1)
print(normal)
print(batched_bernoulli)
```

```python
print(normal.sample(20))
print(bernoulli.sample(3))
print(batched_bernoulli.sample(10))
print(batched_bernoulli.sample(2))
```

```python
plt.hist(exponential.sample(10000),bins=50, density=True)
plt.show()
```

```python
print(bernoulli.prob([-1.0,0.0,0.5,1.]))
print(np.power(0.7,-1.0)*np.power(0.3,2.0))
```
